```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('/WorldCupMatches.csv')


data.head()
```

```python
# Univariate Analysis

# Summary statistics for numerical columns
data.describe()
```

```python
# Numerical Attributes


# Plotting histogram for goals scored by the home team
plt.figure(figsize=(10, 6))
sns.histplot(data['Home Team Goals'], kde=True, bins=30)
plt.title('Distribution of Home Team Goals')
plt.show()

# Plotting histogram for goals scored by the away team
plt.figure(figsize=(10, 6))
sns.histplot(data['Away Team Goals'], kde=True, bins=30)
plt.title('Distribution of Away Team Goals')
plt.show()

# Boxplot to check for outliers in goals scored by both teams
plt.figure(figsize=(10, 6))
sns.boxplot(data[['Home Team Goals', 'Away Team Goals']])
plt.title('Boxplot for Home and Away Team Goals')
plt.show()
```

```python
# Categorical Attributes

# Frequency distribution for categorical variable 'Stage'
plt.figure(figsize=(10, 6))
sns.countplot(data['Stage'])
plt.xticks(rotation=90)
plt.title('Distribution of World Cup Match Stages')
plt.show()
```

```python
# Multivariable Analysis


# Numerical-Numerical Analysis



correlation_matrix = data[['Home Team Goals', 'Away Team Goals', 'Attendance']].corr()

# Heatmap of the correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Scatter plot of goals scored by both teams and attendance
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Home Team Goals', y='Away Team Goals', size='Attendance', data=data, sizes=(20, 200))
plt.title('Scatter Plot of Home vs Away Goals with Attendance Size')
plt.show()
```

```python
# Numerical-Categorical Analysis


# Boxplot of home team goals by stage
plt.figure(figsize=(12, 6))
sns.boxplot(x='Stage', y='Home Team Goals', data=data)
plt.xticks(rotation=90)
plt.title('Home Team Goals by Stage of the Tournament')
plt.show()

# Boxplot of away team goals by stage
plt.figure(figsize=(12, 6))
sns.boxplot(x='Stage', y='Away Team Goals', data=data)
plt.xticks(rotation=90)
plt.title('Away Team Goals by Stage of the Tournament')
plt.show()
```

```python
# Categorical-Categorical Analysis


data['Winning Team'] = np.where(data['Home Team Goals'] > data['Away Team Goals'], 'Home',
                                np.where(data['Home Team Goals'] < data['Away Team Goals'], 'Away', 'Draw'))

# Crosstab of winning team vs stage
crosstab = pd.crosstab(data['Winning Team'], data['Stage'])
print(crosstab)

# Stacked bar plot to visualize the relationship
crosstab.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Winning Team by Tournament Stage')
plt.show()

```

```python
# Univariate Analysis:

# The distribution of goals (both home and away) is right-skewed, meaning most matches have lower goal counts, with fewer matches having large numbers of goals.
# The tournament's most common stages are the group stages, with fewer matches occurring in later knockout rounds.


# Multivariate Analysis:
# There is a weak correlation between the number of goals scored and attendance, indicating that other factors (e.g., the importance of the match) may drive higher attendance.
# The number of goals varies significantly across different stages of the tournament, with earlier stages generally having higher-scoring games.
# Home teams win more often than away teams in all stages of the tournament, but draws are also common, especially in group stages.
```
