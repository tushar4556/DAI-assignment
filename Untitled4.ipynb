```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")


df = pd.read_csv('/content/top_rated_9000_movies_on_TMDB.csv')





df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

#Identifying missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)




numerical_columns = ['vote_average', 'vote_count', 'popularity']



categorical_columns = ['original_language', 'Genres']






# Univariate Analysis for Numerical Variables
def univariate_numerical(df, column):
    plt.figure(figsize=(12,6))
    sns.histplot(df[column], kde=True, color='blue')
    plt.title(f'Distribution of {column}', fontsize=16)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.tight_layout()
    plt.show()

#Univariate analysis on numerical columns
for col in numerical_columns:
    univariate_numerical(df, col)

# Univariate Analysis for Categorical Variables with a limit on categories
def univariate_categorical(df, column, top_n=10):
    plt.figure(figsize=(12,8))

    top_categories = df[column].value_counts().nlargest(top_n).index

    sns.countplot(y=df[column], order=top_categories, palette='coolwarm')
    plt.title(f'Top {top_n} Categories in {column}', fontsize=16)
    plt.xlabel('Count', fontsize=12)
    plt.ylabel(column, fontsize=12)
    plt.tight_layout()
    plt.show()

# Apply univariate analysis on top categories for categorical columns
top_n_categories = 10
for col in categorical_columns:
    univariate_categorical(df, col, top_n=top_n_categories)

# Multivariate Analysis for Numerical Variables (Pairplot)
plt.figure(figsize=(12,6))
sns.pairplot(df[numerical_columns], diag_kind="kde", plot_kws={'alpha':0.5, 's':50, 'edgecolor':'k'})
plt.suptitle("Pairplot for Numerical Variables", fontsize=16)
plt.tight_layout()
plt.show()

# Multivariate Analysis: Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap between Numerical Variables', fontsize=16)
plt.tight_layout()
plt.show()

# Multivariate Analysis for Categorical vs Numerical (Top N Categories)
def boxplot_categorical_numerical(df, categorical_col, numerical_col, top_n=10):
    plt.figure(figsize=(14,8))


    top_categories = df[categorical_col].value_counts().nlargest(top_n).index
    subset_df = df[df[categorical_col].isin(top_categories)]

    sns.boxplot(x=categorical_col, y=numerical_col, data=subset_df, palette='Set3')
    plt.title(f'{numerical_col} vs {categorical_col} (Top {top_n})', fontsize=16)
    plt.xlabel(categorical_col, fontsize=12)
    plt.ylabel(numerical_col, fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Apply boxplot for each categorical vs numerical
for cat_col in categorical_columns:
    for num_col in numerical_columns:
        boxplot_categorical_numerical(df, cat_col, num_col, top_n=top_n_categories)
```

-Popularity and vote count are heavily right-skewed, indicating that a few movies tend to receive disproportionate attention and engagement.

-A popular movie isn't necessarily a highly rated one, as there's little correlation between vote averages and the popularity of a movie. Some widely popular movies may not receive stellar ratings.

-The dominance of English-language films is clear in the dataset, reflecting either a bias in movie production or global consumption of English-language content.

-Action and Adventure films tend to receive higher popularity and vote counts, while Drama and Comedy are the most common genres, even though they don't always achieve the same level of mass popularity.

-The analysis shows that audience engagement (vote counts and popularity) is concentrated in certain genres and languages, with significant disparities across different categories.
